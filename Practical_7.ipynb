### Exercise 21:
Plot a contour plot for the function `z = sin(x) + cos(y)` over a grid of `x` and `y` values ranging from -5 to 5.


```python
# Exercise 21
# prompt: Plot a contour plot for the function z = sin(x) + cos(y) over a grid of x and y values ranging from -5 to 5

import matplotlib.pyplot as plt
import numpy as np
# Create a grid of x and y values
x = np.linspace(-5, 5, 100)
y = np.linspace(-5, 5, 100)
X, Y = np.meshgrid(x, y)

# Calculate the corresponding z values
Z = np.sin(X) + np.cos(Y)

# Create a new figure for the contour plot
plt.figure()

# Create the contour plot
# Use 'levels' to specify the number or locations of contour lines
# Use 'cmap' to specify the colormap (e.g., 'viridis', 'plasma', 'inferno', 'cividis')
contour_plot = plt.contour(X, Y, Z, levels=20, cmap='viridis')

# Add labels and title
plt.xlabel("x")
plt.ylabel("y")
plt.title("Contour Plot of z = sin(x) + cos(y)")

# Add a color bar to show the mapping from contour levels to colors
plt.colorbar(contour_plot, label='z value')

plt.show()
```

### Exercise 26:
Create a 3D plot of the function `z = x^2 + y^2` over a grid of `x` and `y` values.


```python
# Exercise 26
# prompt: Create a 3D plot of the function z = x^2 + y^2 over a grid of x and y values

import matplotlib.pyplot as plt
import numpy as np
from mpl_toolkits.mplot3d import Axes3D

# Create a grid of x and y values
x = np.linspace(-5, 5, 50)
y = np.linspace(-5, 5, 50)
X, Y = np.meshgrid(x, y)

# Calculate the corresponding z values
Z = X**2 + Y**2

# Create a new figure
fig = plt.figure(figsize=(10, 8))

# Add a 3D subplot
ax = fig.add_subplot(111, projection='3d')

# Create the 3D surface plot
ax.plot_surface(X, Y, Z, cmap='viridis') # You can choose a different colormap

# Set labels for the axes
ax.set_xlabel('x')
ax.set_ylabel('y')
ax.set_zlabel('z')

# Set the title of the plot
ax.set_title('3D Plot of z = x^2 + y^2')

# Show the plot
plt.show()
```

```python
# Exercise 34
# prompt: Create a plot with twin y-axes where one axis represents y = x and the other represents y = x^2

import matplotlib.pyplot as plt
import numpy as np
# Exercise 30: Create a plot with twin y-axes where one axis represents y = x and the other represents y = x^2

# Generate x values
x = np.arange(0, 11)

# Create the first figure and the first axis
fig, ax1 = plt.subplots()

# Plot y = x on the first axis
ax1.plot(x, x, 'b-', label='y = x')
ax1.set_xlabel('x')
ax1.set_ylabel('y = x', color='b')
ax1.tick_params('y', colors='b')
ax1.grid(True)

# Create the twin axis
ax2 = ax1.twinx()

# Plot y = x^2 on the twin axis
ax2.plot(x, x**2, 'r-', label='y = x^2')
ax2.set_ylabel('y = x^2', color='r')
ax2.tick_params('y', colors='r')

# Add a title
plt.title("Plot with Twin Y-Axes (y=x and y=x^2)")

# Add legends for both lines
lines, labels = ax1.get_legend_handles_labels()
lines2, labels2 = ax2.get_legend_handles_labels()
ax2.legend(lines + lines2, labels + labels2, loc='upper left')


plt.show()
```

### Exercise 39:
Create a filled contour plot for the function `z = x^2 + y^2` over a grid of `x` and `y` values.


```python
# Exercise 39
# prompt: Create a filled contour plot for the function z = x^2 + y^2 over a grid of x and y values.

import matplotlib.pyplot as plt
import numpy as np
# Create a grid of x and y values
x = np.linspace(-5, 5, 100)
y = np.linspace(-5, 5, 100)
X, Y = np.meshgrid(x, y)

# Calculate the corresponding z values
Z = X**2 + Y**2

# Create a new figure for the filled contour plot
plt.figure()

# Create the filled contour plot
filled_contour = plt.contourf(X, Y, Z, levels=50, cmap='viridis') # Use contourf for filled contours
# You can adjust 'levels' to control the number of filled regions

# Add labels and title
plt.xlabel("x")
plt.ylabel("y")
plt.title("Filled Contour Plot of z = x^2 + y^2")

# Add a color bar to show the mapping from contour levels to colors
plt.colorbar(filled_contour, label='z value')

plt.show()
```

### Exercise 44:

```python
 categories = ['Start', 'Step 1', 'Step 2', 'Step 3', 'End']
values = np.array([50, -20, 30, -15, 45])
cumulative_values = np.cumsum(values)

plt.figure(figsize=(8, 6))
plt.bar(categories, values, bottom=np.append(0, cumulative_values[:-1]), color=['dark blue' if v > 0 else 'blue' for v in values])
plt.xlabel('Steps')
plt.ylabel('Values')
plt.title('Waterfall Chart')
plt.grid(True)
plt.show()
```

### Exercise 46:

### Exercise 49:
Create a radar chart to visualize the strengths of five different attributes.


```python
# Exercise 49
# prompt: Create a radar chart to visualize the strengths of five different attributes

# Data for the radar chart
labels = np.array(['Attribute 1', 'Attribute 2', 'Attribute 3', 'Attribute 4', 'Attribute 5'])
# Replace with your actual attribute strength data (values between 0 and 1 or scaled appropriately)
# Example with sample data:
stats = np.array([0.8, 0.6, 0.9, 0.7, 0.5])

# Number of variables
num_vars = len(labels)

# Compute angle for each axis
angles = np.linspace(0, 2 * np.pi, num_vars, endpoint=False).tolist()

# Complete the circle for plotting
stats = np.concatenate((stats,[stats[0]]))
angles = np.concatenate((angles,[angles[0]]))

# Create the figure and axes
fig, ax = plt.subplots(figsize=(8, 8), subplot_kw=dict(polar=True))

# Plot data
ax.plot(angles, stats, 'o-', linewidth=2)

# Fill area
ax.fill(angles, stats, alpha=0.25)

# Set the grid labels (variable names)
ax.set_thetagrids(np.degrees(angles[:-1]), labels)

# Set the title
ax.set_title("Attribute Strength Radar Chart")

# Set the extent of the radial axis (adjust based on your data range)
ax.set_ylim(0, 1) # Assuming data is scaled between 0 and 1

ax.grid(True)

plt.show()
```

Create a 3D scatter plot with random values for x, y, and z.
